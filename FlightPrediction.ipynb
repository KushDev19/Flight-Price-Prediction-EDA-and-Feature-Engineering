## Flight Price Prediction (EDA + Feature Engineering)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#reading Training data
df_Train = pd.read_excel("Data_Train.xlsx")
#reading Testing data
df_Test = pd.read_excel("Test_set.xlsx")
df_Train
```

```python
df_Test
```

```python
#merging the data
df = pd.concat([df_Train, df_Test])
df
```

```python
df.info()
```

```python
##Feature Engineering Process
##Converting Date_of_journey from object to date time and seperating columns by day, month, year
df['Date'] = df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month'] = df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['Year']= df['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
df.drop('Date_of_Journey',axis=1, inplace=True)
df
```

```python
df['Arrival_Time'] = df['Arrival_Time'].str.split(" ").str[0]
df
```

```python
df['Arrival_Hours'] = df['Arrival_Time'].str.split(':').str[0].astype(int)
df['Arrival_Minutes'] = df['Arrival_Time'].str.split(':').str[1].astype(int)
df['Dep_Hours'] = df['Dep_Time'].str.split(':').str[0].astype(int)
df['Dep_Minutes'] = df['Dep_Time'].str.split(':').str[1].astype(int)
df.head()
```

```python
df.drop(['Dep_Time', 'Arrival_Time'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.drop('Route', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop': 0,'1 stop':1, '2 stops': 2, '3 stops': 3, '4 stops': 4, 'nan': 1})
df['Total_Stops'] = df['Total_Stops'].fillna(1)
df['Total_Stops'].unique()
```

```python
df['Duration_Hours'] = df['Duration'].str.split('h').str[0].str.strip('m').astype(int)
df['Duration_Minutes'] = df['Duration'].str.split('h').str[1].str.strip().str.strip('m').replace({'':0}).fillna(0).astype(int)
```

```python
df.drop('Duration', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Additional_Info'] = df['Additional_Info'].map({'No info':1, 'In-flight meal not included':2,
       'No check-in baggage included':3, '1 Short layover':4, 'No Info':1,
       '1 Long layover':5, 'Change airports':6, 'Business class':7,
       'Red-eye flight':8, '2 Long layover':9})
```

```python
df.head()
```

```python
df.info()
```

```python
df['Airline'] = df['Airline'].map({'IndiGo':1, 'Air India':2, 'Jet Airways':3, 'SpiceJet':4,
       'Multiple carriers':5, 'GoAir':6, 'Vistara':7, 'Air Asia':8,
       'Vistara Premium economy':9, 'Jet Airways Business':10,
       'Multiple carriers Premium economy':11, 'Trujet':12})
```

```python
df['Source'] = df['Source'].map({'Banglore':1, 'Kolkata':2, 'Delhi':3, 'Chennai':4, 'Mumbai':5})
```

```python
df['Destination'] = df['Destination'].map({'New Delhi':1, 'Banglore':2, 'Cochin':3, 'Kolkata':4, 'Delhi':5, 'Hyderabad':6})
```

```python
df.head()
```

```python
df.info()
```

## Feature Engineering

```python
df_Test = df[df['Price'].isnull()]
df_Train = df[~df['Price'].isnull()]
```

```python
X = df_Train.drop('Price', axis=1)
X.isnull().sum()
```

```python
y = df_Train['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Training Model

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import ConstantKernel, RBF

kernel = ConstantKernel(1.0, (1e-3, 1e3)) * RBF(length_scale=1.0, length_scale_bounds=(1e-3, 1e3))
gpr = GaussianProcessRegressor(kernel=kernel, alpha=1e-2, normalize_y=True, n_restarts_optimizer=5, random_state=42)
```

```python
gpr.fit(X_train, y_train)
print("Learned kernel:", gpr.kernel_)

y_pred, y_std = gpr.predict(X_test, return_std=True)

from sklearn.metrics import root_mean_squared_error, r2_score

# Get RMSE directly
rmse = root_mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"RMSE: {rmse:.2f}")
print(f"R²: {r2:.3f}")

```
